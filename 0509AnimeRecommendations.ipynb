#### 애니메이션 추천 시스템 만들기

- **애니메이션 타입 별로 태그처리를 한다. 아까 전에 nan이 있었기 때문에 nan을 제외하고 태그처리하기 위한 타입 리스트를 만든다.**
- **이어서 타입 리스트에 있는 값들을 이름으로 하는 컬럼을 만들어줄거다.그리고 anime\[type\] 값에 맞춰서 O라는 값을 채워줄 것이다.**
- **그리고 type 컬럼은 필요없으니까 버리기~!**

```python
type_list = anime.type.unique()[:-1]
type_list
```

```python
anime[type_list] = "-"

for t in type_list:
    anime[t] = np.where(anime['type'] == t, '○', '-')

anime = anime.drop(columns = ['type', 'episodes', 'members'])
anime
```

- **장르에 대해서도 똑같이 해준다!**
- **NaN을 필터링 해주면서 장르를 뽑아내야 하는데, NaN이 float 형태라서, 뽑아낸 장르의 타입이 0.1과 같은 타입이라면 현재 반복을 건너 뛰고 다음 반복으로 넘어가게 한다. 그 이후 뽑아낸 장르를 장르 딕셔너리의 키 값으로 추가한다.**

```python
genre_dict = {}
genre_list = []

for strs in anime.genre.str.split(', '):
    if type(strs) == type(0.1): # Nan 필터링
        continue
    
    for g in strs:
        genre_dict[g] = 0

for key, val in genre_dict.items():
    genre_list.append(key)

type_list = type_list.tolist()
type_list += genre_list
genre_list
```

```python
anime.head(1)
```

- **장르 리스트를 정리했다. 이제 타입 리스트에 있는 각 타입을 컬럼으로 만들어 채워준 것처럼 장르에 대해서도 똑같이 진핸한다.**
- **np.where의 find 옵션을 써서, 장르 리스트에 있는 g가 해당 데이터 행의 장르 칸에 없다면! -1을 반환한다. 따라서, 데이터 행의 장르 칸에 있으면 anime\[g\]에 O를 넣어주도록 한 것이당.**

```python
for g in genre_list:
    anime[g] = np.where(anime.genre.str.find(g) != -1, '○', '-')

data_anime = anime.drop(['genre'], axis = 1)
data_anime = data_anime.set_index('anime_id')
data_anime
```

- **필요없는 컬럼인 장르 컬럼을 삭제하고, 테이블 인덱스 값은 anime_id로 정의해줬다.**

```python
mergeDF = pd.merge(user_rating, data_anime['name'], on='anime_id')
mergeDF
```

- **사용자 평가 테이블에 유저 아이디 기준으로 영화 제목을 머지해줬다.**

```python
rating_matrix = mergeDF.pivot_table(values = 'rating', index = 'user_id', columns = 'name')
rating_matrix
```

- **머지한 테이블을 가지고 full matrix 구조를 만들어준다. 그럼 이제 각 유저가 평가한 점수들이 각 애니메이션 별로 테이블이 정리되어 구조가 만들어진다. 전체적으로는 NaN밖에 안보이는데, 각각의 유저들이 평가한 애니메이션에는 유저들의 평가 점수가 들어가 있다.**
- **이제는 NaN 값 때문에 연산 오류가 나는 일을 막기 위해, 모두 0으로 바꾸어준다.**

```python
rating_matrix_fillna = rating_matrix.fillna(0)
rating_matrix_fillna
```

```python
# 만약 위 코드가 안되면 이렇게 일일이 할 수도 있지만 좋은 뱅법은 아뉘댱.ㅎ 비효율이에요
# for col, item in rating_matix.iteritems():
#     rating_matrix[col] = item.fillna(0)

# rating_matrix
```

```python
rating_matrix_fillna_np = rating_matrix_fillna.values.T
rating_matrix_fillna_np.shape
```

- **기존 데이터프레임은 69600 x 9922 인데, 특정 애니메이션을 선택할 시 비슷한 애니메이션을 추천해줄 수 있는 시스템을 만들기 위해서는 위에서 만든 매트릭스의 정보 값의 행열 기준을 반전시켜준다. 이전에는 애니메이션을 컬럼으로, 유저 아이디를 행으로 두었는데, Transpose를 통해 애니메이션을 행으로, 유저 아이디를 컬럼으로 설정하는 것이다.**
- **행열 기준 반전을 통해 69600 x 9922 의 데이터프레임이 9922 x 69600이 된 것을 확인할 수 있다.**

```python
rating_matrix_fillna_np
```

#### 추천 시스템 학습
- **TruncatedSVD를 활용해서 학습을 시킨다. full matrix 전체 테이블을 가지고 사용자 기반 협업 필터링을 통해 추천 값을 리스트로 뽑아 보여줄 수도 있지만 현재 테이블이 개인적으로 프로젝트를 하기에는 메모리를 너무 크게 사용한다. 따라서, 행렬 분해 방식을 통해 애니메이션 별로 연관성을 학습하고, 연산할 수 있는 테이블을 통해 값을 뽑아내는 방식을 쓰기 위해 TruncatedSVD를 사용하기로!**

```python
user_count, ani_count = rating_matrix_fillna_np.shape

from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components = 12)
matrix = SVD.fit_transform(rating_matrix_fillna_np)
matrix.shape
```

```python
corr= np.corrcoef(matrix)
corr.shape
```

- **TruncatedSVD를 통해 뽑아낸 정보를 가지고 피어슨 상관계수를 가져와서 계산한다. 상관계수를 기준으로 특정 애니메이션과 가장 비슷한 애니메이션을 찾아올 수 있도록 해주는 것이다.**

```python
title_list = list(rating_matrix_fillna.columns)
title_list
```

#### 추천시스템 완성
- **검색할 수 있는 애니메이션 이름을 리스트로 뽑는다. 이제 애니메이션 검색하면 이 리스트에서 애니메이션 매칭하는거다.**
1. **사용자로부터 애니메이션 제목을 입력값으로 받고, 애니메이션 제목 리스트에서 입력값으로 받은 애니메이션 제목의 인덱스를 찾는다.**
2. **인덱스 값을 기준으로 입력받은 애니메이션의 피어슨 상관계수 값을 계산한다.**
3. **피어슨 상관계수 값이 90% 이상인 애니메이션을 최대 10개를 뽑아서 테이블 형태로 보여준다.**

```python
selected_title = input("Enter the Title of Animation : ")

idx = title_list.index(selected_title)
similarity = corr[idx]

idx = 0
idx_dict = {}

for val in similarity:
    if val >= 0.9:
        idx_dict[val] = title_list[idx]
    idx += 1

ani_names = []
limit = 10
idx = 0
for key, val in sorted (idx_dict.items(), reverse = True):
    if limit < idx:
        break
    if idx == 0:
        idx += 1
        continue
    
    ani_names.append(val)
    idx += 1
    
answer_table = data_anime[data_anime.name.isin(ani_names)]
answer_table
```
